# Visualizing the Distribution of a Categorical Variable (e.g., Gender):

```python
# Counting the occurrences of each category in the 'Gender' column
gender_counts = data['Gender'].value_counts()

plt.bar(gender_counts.index, gender_counts.values)

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Gender')

plt.show()
```

# Visualizing the Distribution of a Continuous Variable (e.g., Age):


```python
import seaborn as sns

# Creating a histogram
sns.histplot(data=data, x='Age', bins=20, kde=True)

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')

plt.show()
```

# Categorical variable visualisation using barchart

```python
data['Region'].unique()
```

```python
categories = data['Region'].unique()

# Create separate bar charts for each category
for category in categories:
    subset_data = data[data['Region'] == category]
    category_counts = subset_data['Gender'].value_counts()
    plt.figure(figsize=(5,5))
    plt.bar(category_counts.index, category_counts.values)
    plt.xlabel('Gender')
    plt.ylabel('Count')
    plt.title(f'Distribution of Gender in {category}')
    plt.show()
```

# Define the subsets of data you want to visualize separately


```python
subset1 = data[data['Marital_status'] == 'Single']
subset2 = data[data['Marital_status'] == 'Married']

# Creating separate histograms for each subset
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.histplot(data=subset1, x='Age', bins=20, kde=True)
plt.title('Distribution of Age for Single Individuals')

plt.subplot(1, 2, 2)
sns.histplot(data=subset2, x='Age', bins=20, kde=True)
plt.title('Distribution of Age for Married Individuals')

plt.tight_layout()
plt.show()
```
